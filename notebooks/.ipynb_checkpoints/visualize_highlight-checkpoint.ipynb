```python
from IPython.core.display import display, HTML
import pickle as pkl
import html
```

```python
idx = 0
```

```python
viz_folder = '../viz_distributions'
```

```python
rbg = '135,206,250'
max_alpha = 0.8
def get_html_highlight(text, span2weight):
    words = text.split()
    highlighted_text = ""
    for span, weight in span2weight.items():
        phrase = ' '.join(words[span[0]:span[1]+1])
        highlighted_text += f'<span style="background-color:rgba({rbg},' + str(weight / max_alpha) + ');">' + html.escape(phrase)+ '</span>'
    return highlighted_text
```

```python
extreme_results = pkl.load(open(f'{viz_folder}/{idx}.pkl'.format(idx=idx), 'rb'))
note = extreme_results['note']

for prefix in ['neg', 'pos']:
    d = extreme_results[prefix + '2score']
    sorted_texts = sorted(d, key=lambda t: d[t], reverse=True)
    for t in sorted_texts[:10]:
        span2weight = extreme_results[prefix + '2highlight'][t]
        h = get_html_highlight(t, span2weight)
        display(HTML(h))
    print('=========')
idx += 1
```
